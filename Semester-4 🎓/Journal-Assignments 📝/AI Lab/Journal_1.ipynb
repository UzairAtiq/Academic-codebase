# Class



## Task 1 & 2

```python
class student : 

  def __init__(self,id,name,subj1,subj2,subj3):
    self.id = id
    self.name = name
    self.subj1 = subj1
    self.subj2 = subj2
    self.subj3 = subj3

  def calculateTotalMarks(self):
    return self.subj1 + self.subj2 + self.subj3

  def calculateAverageMarks(self):
    totalMarks = self.calculateTotalMarks()
    return totalMarks / 3

  def assignGrade(self):
    average = self.calculateAverageMarks()
    if average >= 85:
      return "A"
    elif average >= 70:
      return "B"
    elif average >= 50:
      return "C"
    else:
      return "F"

  def display(self) :
    print(self.id)
    print(self.name)
    print(self.subj1)
```

```python
#For Storing Objects

Students = []

for i in range(2) :
  try:
    idTest = int(input("Enter ID"))
  
  except:
    print("Invalid Input")

  nameTest = input(f"Please Enter Student {i} Name: ")
  try:
    subj1 = int(input("Enter Subj1"))
    if subj1 > 100 or subj1 < 0 :
      raise ValueError("Value Out of Bound")
    else :
      pass
      
    subj2 = int(input("Enter Subj2"))
    if subj2 > 100 or subj2 < 0 :
      raise ValueError("Value Out of Bound")
    else :
      pass
     
    subj3 = int(input("Enter Subj3"))
    if subj3 > 100 or subj3 < 0 :
      raise ValueError("Value Out of Bound")
    else :
      pass
  
  except:
    print("Enter A Integer")
  
  s = student(idTest,nameTest,subj1,subj2,subj3)
  Students.append(s)
```

# Task 3 

```python
for i in range(2) :
  print(f"Record {i} ")
  print("Total: ", Students[i].calculateTotalMarks())
  print("Average: ", Students[i].calculateAverageMarks())
  print("Grade: ", Students[i].assignGrade())
```

# Task 4

```python
with open("results.txt", "w") as file:
    for student in Students:
        
        totalMarks = student.calculateTotalMarks()
        averageMarks = student.calculateAverageMarks()
        grade = student.assignGrade()

        file.write(f"Student ID: {student.id}\n")
        file.write(f"Name: {student.name}\n")
        file.write(f"Total Marks: {totalMarks}\n")
        file.write(f"Average Marks: {averageMarks}\n")
        file.write(f"Grade: {grade}\n")
        file.write("-" * 30 + "\n")
```

# Task 6

```python
averages = [student.calculateAverageMarks() for student in Students]

highestAverage = max(averages)
lowestAverage = min(averages)
classAverage = sum(averages) / len(averages)

print("Highest Average Score:", highestAverage)
print("Lowest Average Score:", lowestAverage)
print("Class Average:", classAverage)
```
