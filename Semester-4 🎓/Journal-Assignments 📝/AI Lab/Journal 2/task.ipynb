```python
import math

class basic_calc:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    
    def addition(self):
        return self.num1 + self.num2
    
    def subtraction(self):
        return self.num1 - self.num2
    
    def multiplication(self):
        return self.num1 * self.num2
    
    def division(self):
        if self.num2 != 0:
            return self.num1 / self.num2
        return "Cannot divide by zero"

class s_calc(basic_calc):
    def factorial(self, n):
        return math.factorial(n)
    
    def xPowerY(self, x, y):
        return x ** y
    
    def log(self, value, base=10):
        return math.log(value, base)
    
    def ln(self, value):
        return math.log(value)

calc = s_calc(10, 5)
print("Addition:", calc.addition())
print("Subtraction:", calc.subtraction())
print("Multiplication:", calc.multiplication())
print("Division:", calc.division())
print("Factorial of 5:", calc.factorial(5))
print("10 power 2:", calc.xPowerY(10, 2))
print("Log base 10 of 100:", calc.log(100))
print("Natural log of 10:", calc.ln(10))
```

```python
def medicalDiagnosis():
    print("Enter your symptoms (yes/no):")
    fever = input("Do you have fever? ").lower()
    painAbdomen = input("Do you have pain in abdomen (especially iliac fossa)? ").lower()
    vomiting = input("Do you have vomiting? ").lower()
    cough = input("Do you have cough? ").lower()
    chestPain = input("Do you have chest pain? ").lower()
    
    print("\nEnter test results:")
    tlcHigh = input("Is TLC high? ").lower()
    neutrophilsHigh = input("Are neutrophils high? ").lower()
    esrHigh = input("Is ESR high? ").lower()
    
    if fever == "yes" and painAbdomen == "yes" and vomiting == "yes":
        if tlcHigh == "yes" and neutrophilsHigh == "yes" and esrHigh == "yes":
            print("\nDiagnosis: Acute Appendicitis")
            print("Treatment: Surgery")
            return
    
    if fever == "yes" and cough == "yes" and chestPain == "yes":
        if tlcHigh == "yes" and neutrophilsHigh == "yes" and esrHigh == "yes":
            xray = input("Does X-ray show pneumonic patch? ").lower()
            if xray == "yes":
                print("\nDiagnosis: Pneumonia")
                print("Treatment: Antibiotics")
                return
    
    if fever == "yes" and cough == "yes":
        throatRed = input("Are tonsils red and enlarged? ").lower()
        pusTonsils = input("Is there pus in tonsils? ").lower()
        if throatRed == "yes" and pusTonsils == "yes":
            print("\nDiagnosis: Acute Tonsilitis")
            print("Treatment: Anti-allergic, paracetamol. If not gone, add antibiotics orally. If not gone, add antibiotics IV")
            return
    
    print("\nCannot determine diagnosis based on provided symptoms and tests")

medicalDiagnosis()
```
